### 1.1 (20 points)

Consider $1$-dimensional points $x_i\in \mathbb{R}$ and prove or disprove that the following two distance functions are metrics:

$$
d_1(x_i,x_j) = |x_i^2 - y_i^2|,
$$

and
$$
d_2(x_i,x_j) = |x_i^3 - y_i^3|.
$$


### Proof 1 :

#### 1. Non-negativity : d(x,y) >= 0

Since the given distance function is computed as the absolute value of the difference operation, it is always greater than or equal to zero. <br>

#### 2. Isolation : d(x,y) = 0 <=> x = y

#### a) x = y => d(x,y) = 0

d(x,y) = d(x,x) = |x^2 - x^2| = 0 <br>

#### b) d(x,y) = 0 => x = y

   d(x,y)   = 0 <br>
|x^2 - y^2| = 0 <br> 
 x^2 - y^2  = 0 <br>
    x^2     = y^2 <br>
  (+/-) x  != (+/-) y <br>

### Isolation property fails. Hence it is not a distance metric.

### Proof 2 :

#### 1. Non-negativity : d(x,y) >= 0

Since the given distance function is computed as the absolute value of the difference operation, it is always greater than or equal to zero. <br>

#### 2. Isolation : d(x,y) = 0 <-> x = y

#### a) x = y -> d(x,y) = 0

d(x,y) = d(x,x) = |x^3 - x^3| = 0

#### b) d(x,y) = 0 -> x = y

   d(x,y)   = 0 <br>
|x^3 - y^3| = 0 <br>
 x^3 - y^3  = 0 <br>
    x^3     = y^3 <br>
     x      = y <br>

#### 3. Symmetry : d(x,y) = d(y,x)

d(x,y) = |x^3 - y^3| <br>
       = |-y^3 + x^3| <br>
       = |(-1) (-y^3 + x^3)| <br>
       = |y^3 - x^3|<br>
       = d(y,x) <br>

#### 4. Triangle inequality : d(x,y) <= d(x,z) + d(z,y)

d(x,y) = |x^3 - y^3| <br>
       = |x^3 - y^3 + 0| <br>
       = |x^3 - y^3 + z^3 - z^3| <br>
       = |(x^3 - z^3) + (z^3 - y^3)| <br>
      <= |x^3 - z^3| + |z^3 - y^3| <br>
      
(Since when both the terms are positive or negative,  |x + y| = |x| + |y| and when the terms have opposite signs, 
|x + y| < |x| + |y|) <br>

### Since it satisfies all 4 properties, it is a distance metric.

#### 1.2.2 (10 points)
Construct a graph $G=(V,E)$, where a point $x_i$ is represented
  by node $v_i\in V$. For every pair of nodes $v_i,v_j$, there exists an
  undirected edge in $G$ with weight $w_{ij}=d_R(x_i,x_j)$. We define the
  distance function between two nodes $v_i$ and $v_j$, denoted by
  $d_G(v_i,v_j)$, be the **weight of the shortest path between $v_i$ and $v_j$** in graph $G$. Prove or disprove that $d_G$ is a metric. 


### Proof :

#### 1. Non-negativity : d(x,y) >= 0

Given that, d(x,y) is the weight of the shortest path between x and y. <br>
Since the weights of all the paths lie between (0,∞), the summation of the weights of all the edges in the path also lies between (0,∞) <br>
So, <br>
d(x,y) >= 0 <br>

#### 2. Isolation : d(x,y) = 0 <-> x = y

#### a) x = y -> d(x,y) = 0

d(x,y) = d(x,x) = 0 (This is obviously the shortest path between x and x) <br>

#### b) d(x,y) = 0 -> x = y

d(x,y) = 0 <br>
Given that d(x,y) = 0 only when x = y and is (0,∞) otherwise. <br>
This implies,  <br>
     x = y <br>
   
#### 3. Symmetry : d(x,y) = d(y,x)

Given that the weight of each edge is calculated only once from the vertex with the smaller index to the vertex with higher index. <br>
Since the edges are undirected, the same weight applies in the opposite direction. <br>
So, the shortest path from x to y is also the shortest path from y to x since the summation of the weights of all the edges in the path is the same in both directions. <br>
So, <br>
d(x,y) = d(y,x)



#### 4. Triangle inequality : d(x,y) <= d(x,z) + d(z,y)

Since d(x,y) is the shortest distance between x and y, any other path between x and y such as d(x,z) + d(z,y) will always have a distance greater than or equal to it.
So, <br>
d(x,y) <= d(x,z) + d(z,y)

### Since it satisfies all 4 properties, it is a distance metric. 

## **2. Experiments**

This is a python program that generates a random sequence of symbols (letters and numbers) of size $n$

```python
import random
import string

def String(n):

    S = ''.join(random.choice(string.ascii_uppercase + string.digits) for _ in range(n))
    return S
```

### 2.1 (20 points)

Implement a routine (let's call it **Permutation**) that selects a random sample $k$ of the elements in string $S$ without replacement as follows:

1. Generate a random permutation $\pi$ of the elements in $S$. Using the linear-time algorithm we discussed in class.
2. Pick the first $k$ elements in the permuted version of $S$. 



```python
def Permutation(S, k):
    
    S = list(S)
    for i in range(1, len(S)):
        j = random.randint(0, i-1)
        S[i], S[j] = S[j], S[i]
    print(S[:k])
```

### 2.2 (20 points)
Implement reservoir sampling (let's call this routine **Reservoir**) as we discussed it in class.

```python
def Reservoir(S,k):
    
    S = list(S)
    R = S[:k] 
    for i in range(k, len(S)):
        rand = random.randint(0, i)
        if(rand < k):
            R[rand] = S[i] 
    print(R)
```
